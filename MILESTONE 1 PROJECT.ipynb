# Hypothesis testing

```python
import numpy as np
from scipy import stats

# Given values
p1 = 0.27975  # Resort hotel cancellation proportion
p2 = 0.417082  # City hotel cancellation proportion
n1 = 39000  # Total bookings for resort hotel
n2 = 73000  # Total bookings for city hotel

# Pooled proportion
pooled_p = (p1 * n1 + p2 * n2) / (n1 + n2)

# Z-statistic calculation
z_stat = (p1 - p2) / np.sqrt(pooled_p * (1 - pooled_p) * (1/n1 + 1/n2))

# P-value calculation (two-tailed test)
p_value = 2 * (1 - stats.norm.cdf(abs(z_stat)))

# Output results
print(f"Z-statistic: {z_stat}")
print(f"P-value: {p_value}")

# Interpretation
alpha = 0.05  # Significance level
if p_value < alpha:
    print("Reject the null hypothesis: Significant difference between cancellation proportions")
else:
    print("Fail to reject the null hypothesis: No significant difference between cancellation proportions")
```
